# Solved midpoints?
> Itsik claims to have solved the midpoint problem analytically. Does it work?

We need to find $\theta_m$ such that
$$
\begin{align}
\cos(2\theta_2)\cos^2(\theta_1 + \theta_m) - \cos(2\theta_1)\cos^2(\theta_2 + \theta_m) &= 0\\
(2\cos^2(\theta_2) - 1)\cos^2(\theta_1 + \theta_m) - (2\cos^2(\theta_1 - 1))\cos^2(\theta_2 + \theta_m) &= 0\\
(2\cos^2(\theta_2) - 1)(\cos(\theta_1)\cos(\theta_m) - \sin(\theta_1)\sin(\theta_m))^2 - (2\cos^2(\theta_1 - 1))(\cos(\theta_1)\cos(\theta_m) - \sin(\theta_1)\sin(\theta_m))^2 &= 0
\end{align}
$$

If we define
$$
\begin{align}
    U &:= (2\cos^2(\theta_2) - 1)\cos^2(\theta_1) - (2\cos^2(\theta_1))\cos^2(\theta_2)\\
    &= \cos^2(\theta_2) - \cos^2(\theta_2)\\
    V &:= (2 \cos^2(\theta_2) -1)2\cos(\theta_1)\sin(\theta_1) - (2\cos^2(\theta_1) -1) 2\cos(\theta_2)\sin(\theta_2)\\
    &= \cos(2\theta_2)\sin(2\theta_1) - \cos(2\theta_1)\sin(2\theta_2)\\
    &= \sin(2\theta_1 - 2\theta_2) \\
    W &:= (2 \cos^2(\theta_2) -1)\sin^2(\theta_1) - (2 \cos^2(\theta_1) -1)\sin^2(\theta_2)\\
    &= (1-2 \sin^2(\theta_2))\sin^2(\theta_1) - (1-2\sin^2(\theta_1))\sin^2(\theta_2)\\
    &= \sin^2(\theta_1) - \sin^2(\theta_2)\\
    &= \cos^2(\theta_2) - \cos^2(\theta_1)
\end{align}
$$
then we can express this equation as $U\cot(\theta_m)^2 + V\cot(\theta_m) + W = 0$, a quadratic equation in $\cot(\theta_m)$.

Notice that both $U$ and $W$ are $\cos^2(\theta_2) - \cos^2(\theta_1)$. We can further simplify $U, V,$ and $W$ by dividing each by this expression:
$$
\begin{align}
U &= \frac{\cos^2(\theta_2) - \cos^2(\theta_1)}{\cos^2(\theta_2) - \cos^2(\theta_1)}\\
&= 1\\
V &= \frac{\sin(2\theta_1 - 2\theta_2)}{\cos^2(\theta_2) - \cos^2(\theta_1)}\\
&= \frac{\sin(2\theta_1 - 2\theta_2)}{(\cos(\theta_2) + \cos(\theta_1))(\cos(\theta_2) - \cos(\theta_1))}\\
&= \frac{\sin(2\theta_1 - 2\theta_2)}{-4\cos(\frac{\theta_1 + \theta_2}{2})\cos(\frac{\theta_2 - \theta_1}{2})\sin(\frac{\theta_1 + \theta_2}{2})\sin(\frac{\theta_2 - \theta_1}{2})}\\
&= -\frac{\sin(2\theta_1 - 2\theta_2)}{\sin(\theta_2 + \theta_1)\sin(\theta_2 - \theta_1)}\\
W &= \frac{\cos^2(\theta_2) - \cos^2(\theta_1)}{\cos^2(\theta_2) - \cos^2(\theta_1)}\\
&= 1
\end{align}
$$

Therefore we have the expression
$$
\cot^2(\theta_m) + \frac{\sin(2\theta_2 - 2\theta_1)}{\sin(\theta_1 + \theta_2)\sin(\theta_1 - \theta_2)}\cot(\theta_m) + 1 = 0
$$

By substituting $X := \cot(\theta_m)$, we get a quadratic equation:

$$
UX^2 + VX + W = 0
$$

The roots of this occur at 
$$
\frac{-V \pm \sqrt{V^2 - 4}}{2}
$$

Specifically, we have
$$
\theta_m = \begin{cases}
\theta_1 & \text{ if } \theta_1 = \theta_2\\
\cot^{-1}\left( \frac{-V - \sqrt{V^2 - 4}}{2} \right) & \text{ if } \theta_1 < \pi - \theta_2\\
\cot^{-1}\left( \frac{-V + \sqrt{V^2 - 4}}{2} \right) & \text{ if } \theta_1 > \pi - \theta_2\\
\end{cases}
$$

# Numerical investigation

```python
# New quadratic:

# We can derive this:
# cot^2(M) + [sin(2A - 2B) / sin(B+A)sin(A-B)]cot(M) + 1 = 0

import numpy as np
import matplotlib.pyplot as plt


def check_is_midpoint(a, b, m):
    bool1 = a <= m <= b or b <= m <= a
    bool2 = np.isclose(np.cos(2 * a) * np.cos(b + m) ** 2, np.cos(2 * b) * np.cos(a + m) ** 2)
    return bool1 and bool2


eps = 1e-4
# thetas = np.arange(np.pi / 4 + eps, 3 * np.pi / 4 - eps, 0.01)
thetas = np.arange(0, np.pi, 0.01)
# thetas = np.arange(0, 2 * np.pi, 0.01)

my_grid1 = np.ones((len(thetas), len(thetas))) * np.nan
my_grid2 = np.ones((len(thetas), len(thetas))) * np.nan
for i, a in enumerate(thetas):
    for j, b in enumerate(thetas):
        if np.isclose(a, b):
            my_grid1[i, j] = a
            my_grid2[i, j] = a
            continue
        # Let T := cot(M)
        # Then we have the quadratic equation:
        # T^2 + [sin(2A - 2B) / sin(B+A)sin(A-B)]T + 1 = 0
        sols = np.roots(
            [
                1,
                np.sin(2 * a - 2 * b) / (np.sin(a + b) * np.sin(b - a)),
                1,
            ]
        )
        try:
            sols = np.arctan2(1, sols[np.isreal(sols)]) % np.pi
            if len(sols) > 0:
                my_grid1[i, j] = sols[0]
            if len(sols) > 1:
                my_grid2[i, j] = sols[1]
                if not check_is_midpoint(a, b, sols[1]):
                    print(f"No midpoint found for {a/np.pi:.3f}, {sols[1]/np.pi:.3f}, {b/np.pi:.3f}")
        except:
            pass


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
ax1.imshow(my_grid1)
ax2.imshow(my_grid2)
plt.show()
```

```python
# Now with analytic solution

# New quadratic:

# We can derive this:
# cot^2(M) + [sin(2A - 2B) / sin(B+A)sin(A-B)]cot(M) + 1 = 0

import numpy as np
import matplotlib.pyplot as plt


def check_is_midpoint(a, b, m):
    bool1 = a <= m <= b or b <= m <= a
    bool2 = np.isclose(np.cos(2 * a) * np.cos(b + m) ** 2, np.cos(2 * b) * np.cos(a + m) ** 2)
    return bool1 and bool2


eps = 1e-4
thetas = np.arange(np.pi / 4 + eps, 3 * np.pi / 4 - eps, 0.01)
# thetas = np.arange(0, 2 * np.pi, 0.01)

my_grid1 = np.ones((len(thetas), len(thetas))) * np.nan
for i, a in enumerate(thetas):
    for j, b in enumerate(thetas):
        if np.isclose(a, b):
            m = a

        else:
            v = np.sin(2 * a - 2 * b) / (np.sin(a + b) * np.sin(b - a))
            coef = -1 if a < np.pi - b else 1
            sol = (-v + coef * np.sqrt(v**2 - 4)) / 2
            m = np.arctan2(1, sol) % np.pi  # sol = cot(m)

        my_grid1[i, j] = m

        if not check_is_midpoint(a, b, m):
            print(f"No midpoint found for {a/np.pi:.3f}, {m/np.pi:.3f}, {b/np.pi:.3f}")


plt.matshow(my_grid1)
```
