# Fix high-dim gaussians
> We get a lot of off-manifold points. Maybe it's because the norms get too big?

```python
%load_ext autoreload
%autoreload 2
```

```python
import matplotlib.pyplot as plt
import numpy as np

from src.hyperdt.toy_data import wrapped_normal_mixture
```

```python
norms = []
dims = []
for dim in range(2,16):
    for _ in range(10):
        X, y = wrapped_normal_mixture(n_dim=dim, num_points=1000, num_classes=2)
        norms.append(np.linalg.norm(X[:,1:], axis=1).mean())
        dims.append(dim)

plt.scatter(dims, norms)
```

```python
# Let's try the same thing dividing out by square of dimensionality

norms = []
dims = []
for dim in range(2,16):
    for _ in range(10):
        X, y = wrapped_normal_mixture(n_dim=dim, num_points=1000, num_classes=2, noise_std = 1 / dim)
        norms.append(np.linalg.norm(X[:,1:], axis=1).mean())
        dims.append(dim)

plt.scatter(dims, norms)
```
